```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
print("Running Diabetes Model")

df = pd.read_csv("NBD.csv")

X = df.drop('diabetes', axis=1)
y = df['diabetes']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = Sequential()
model.add(Dense(64, activation='relu', input_dim=X.shape[1]))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=50, batch_size=10)

loss, acc = model.evaluate(X_test, y_test)

print("Diabetes Model Accuracy:", acc)
```

```python
print("Running Lung Cancer Model")

lung = pd.read_csv("lungcancer.csv")

lung['LUNG_CANCER'] = lung['LUNG_CANCER'].map({'YES':1,'NO':0})
lung['GENDER'] = lung['GENDER'].map({'M':1,'F':0})

X2 = lung.drop("LUNG_CANCER", axis=1)
y2 = lung["LUNG_CANCER"]

X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2)

model2 = Sequential()
model2.add(Dense(64, activation='relu', input_dim=X2.shape[1]))
model2.add(Dense(32, activation='relu'))
model2.add(Dense(1, activation='sigmoid'))

model2.compile(optimizer='adam',
               loss='binary_crossentropy',
               metrics=['accuracy'])

model2.fit(X2_train, y2_train, epochs=50, batch_size=10)

loss, acc = model2.evaluate(X2_test, y2_test)

print("Lung Cancer Model Accuracy:", acc)
```
